```python
#Convert GAP polygons to raster
arcpy.env.overwriteOutput = True
with arcpy.EnvManager(snapRaster="OGDP"):
    arcpy.conversion.PolygonToRaster(
        in_features="oil_states_GAP12_albers",
        value_field="BndryExten",
        out_rasterdataset=r"C:\Users\MitchellChandler\OneDrive - THE WILDERNESS SOCIETY\OGDP model\_Data\gis_data\explore_data\explore_data.gdb\oil_states_GAP_raster",
        cell_assignment="MAXIMUM_AREA",
        priority_field="GIS_Acres",
        cellsize=10000,
        build_rat="BUILD"
    )
```

```python
#Convert grid cells to points
arcpy.env.overwriteOutput = True

#OGDP
arcpy.conversion.RasterToPoint(
    in_raster="OGDP",
    out_point_features=r"C:\Users\MitchellChandler\OneDrive - THE WILDERNESS SOCIETY\OGDP model\_Data\gis_data\explore_data\explore_data.gdb\ogdp_points",
    raster_field="Value"
)

#GAP 1+2
arcpy.conversion.RasterToPoint(
    in_raster="oil_states_GAP_raster",
    out_point_features=r"C:\Users\MitchellChandler\OneDrive - THE WILDERNESS SOCIETY\OGDP model\_Data\gis_data\explore_data\explore_data.gdb\gap_raster_points",
    raster_field="Value"
)
```

```python
#GAP and non-GAP points
arcpy.env.overwriteOutput = True

#Erase OGDP points where there is GAP land to obtain non-GAP points 
arcpy.analysis.PairwiseErase(
    in_features="ogdp_points",
    erase_features="gap_raster_points",
    out_feature_class=r"C:\Users\MitchellChandler\OneDrive - THE WILDERNESS SOCIETY\OGDP model\_Data\gis_data\explore_data\explore_data.gdb\nongap_points",
    cluster_tolerance=None
)

#Erase OGDP non-GAP points to obtain GAP points
arcpy.analysis.PairwiseErase(
    in_features="ogdp_points",
    erase_features="nongap_points",
    out_feature_class=r"C:\Users\MitchellChandler\OneDrive - THE WILDERNESS SOCIETY\OGDP model\_Data\gis_data\explore_data\explore_data.gdb\gap_points",
    cluster_tolerance=None
)
```

```python
#Export as netCDF
arcpy.env.overwriteOutput = True

#GAP
arcpy.md.FeatureToNetCDF(
    in_features="gap_points",
    fields_to_variables="grid_code grid_code #",
    out_netCDF_file=r"C:\Users\MitchellChandler\OneDrive - THE WILDERNESS SOCIETY\OGDP model\_Data\scripts\results\ogdp50_gap.nc",
    fields_to_dimensions=None
)

#non-GAP
arcpy.md.FeatureToNetCDF(
    in_features="nongap_points",
    fields_to_variables="grid_code grid_code #",
    out_netCDF_file=r"C:\Users\MitchellChandler\OneDrive - THE WILDERNESS SOCIETY\OGDP model\_Data\scripts\results\ogdp50_nongap.nc",
    fields_to_dimensions=None
)
```
